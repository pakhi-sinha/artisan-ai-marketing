```python
from google.cloud import vision

client = vision.ImageAnnotatorClient()

# Use the uploaded image filename
file_name = "handwovenbasket.jpg"  # replace with your uploaded file name

with open(file_name, "rb") as image_file:
    content = image_file.read()

image = vision.Image(content=content)

response = client.label_detection(image=image)
labels = response.label_annotations

print("Labels detected:")
for label in labels:
    print(label.description, "-", label.score)
```

```python
# Convert labels to hashtags
hashtags = []
for label in labels:
    tag = label.description.replace(" ", "").lower()
    hashtags.append(f"#{tag}")

print("Generated Hashtags for Social Media:")
print(" ".join(hashtags))
```

```python
# Generate a simple description using top labels
top_labels = [label.description for label in labels[:3]]  # take top 3 labels
description = f"This handmade item is related to {', '.join(top_labels)}."

print("Generated Description:")
print(description)
```

```python
from google.cloud import translate_v2 as translate

# Initialize Translation client
translate_client = translate.Client()

# Translate into Hindi and Bengali
target_languages = ["hi", "bn"]

print("\nTranslations:")
for lang in target_languages:
    result = translate_client.translate(description, target_language=lang)
    print(f"{lang.upper()}:", result["translatedText"])
```

```python
from google.cloud import texttospeech, speech

tts_client = texttospeech.TextToSpeechClient()
stt_client = speech.SpeechClient()
```

```python
# ---------------------------
# Text-to-Speech for your craft
# ---------------------------
marketing_text = "This handmade craft features a wooden handmade basket. Perfect for gifts and home decoration."

synthesis_input = texttospeech.SynthesisInput(text=marketing_text)

voice = texttospeech.VoiceSelectionParams(
    language_code="en-US",
    ssml_gender=texttospeech.SsmlVoiceGender.FEMALE
)

audio_config = texttospeech.AudioConfig(audio_encoding=texttospeech.AudioEncoding.MP3)

response = tts_client.synthesize_speech(
    input=synthesis_input,
    voice=voice,
    audio_config=audio_config
)

tts_file = "handmade_basket.mp3"
with open(tts_file, "wb") as out:
    out.write(response.audio_content)

print(f"✅ TTS audio saved: {tts_file}")

# Play the audio in Colab
from IPython.display import Audio
Audio(tts_file, autoplay=True)
```

```python
import pydub

# Convert MP3 to WAV
audio = pydub.AudioSegment.from_mp3(tts_file)
wav_file = "handmade_basket.wav"
audio.export(wav_file, format="wav")

with open(wav_file, "rb") as f:
    content = f.read()

audio_input = speech.RecognitionAudio(content=content)
config = speech.RecognitionConfig(
    encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
    sample_rate_hertz=audio.frame_rate,
    language_code="en-US"
)

stt_response = stt_client.recognize(config=config, audio=audio_input)

for result in stt_response.results:
    print("🎤 STT transcription:", result.alternatives[0].transcript)
```
